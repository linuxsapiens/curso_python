```python
# Gráfico lineal, es posible personzalizar forma, tamaño, color, las líneas y marcadores
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples)
```

```python
# Podriamos mejorar la descripción mostrando el año ( supuesto ) usando dos argumentos para plot
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]

plt.plot(years, yield_apples);
```

```python
# Podemos agregar etiquetas a los ejes para mostrar lo que representan 
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
# Si quisiéramos mostrar varias líneas, como manzanas y naranjas
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]

plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
# Para identificar qué representa cada línea se puede incluir una leyenda, así como un título
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
# Ya solo faltaría agregar marcadores, en matplotlib existen distintos como círculo, cruz, cuadrado, diamante
# La lista completa se puede consultar en https://jovian.ai/outlink?url=https%3A%2F%2Fmatplotlib.org%2F3.1.1%2Fapi%2Fmarkers_api.html
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
# La función plot soporta varios argumentos para dar estilo a las líneas y marcadores
#    color o c – Ajusta el color de la línea (supported colors)
#    linestyle o ls – Elije entre una línea sólida o discontinua
#    linewidth o lw – Establece el ancho de una línea
#    markersize o ms – Fija el tamaño de los marcadores
#    markeredgecolor o mec – Establecer el color del borde para los marcadores
#    markeredgewidth o mew – Establecer el ancho del borde para los marcadores
#    markerfacecolor o mfc – Establecer el color de relleno para los marcadores
#    alpha – Opacidad de la gráfica
```

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
# El argumento fmt proporciona una abreviatura para especificar la forma del marcador, el estilo de línea y el color de línea.
# Puede proporcionarlo como tercer argumento para plt.plot.
fmt = '[marker][line][color]'

plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
# Se puede usar la función "figure" para cambiar el tamaño de la figura, antes de usar figure
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

```python
# Gráficos de dispersión con MatPlotLib
# En un diagrama de dispersión, los valores de 2 variables se representan como puntos en una cruadrícula bidimensional
# Además, también se pude utilizar una tercera variable para especificiar el tamaño o colo de los puntos. 
```

```python
# Seaborn incluye dataframes o datasets de muestra que podemos usar
# "Iris flower dataset" proporciona medidas de muestra de sépalos y pétalos para tres especies de flores.
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")

flowers_df
```

```python
# Inspeccionar las especies
flowers_df.species.unique()
```

```python
# Si quisieramos ver la relación entre el largo y ancho del sépalo
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
# Parece muy claro, no es evidente que hubiera una relación entre ellos
# Pero se puede usar un diagrama de dispersión para ver como varían el largo y ancho Usando scatterplot
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

```python
# Para distinguir entre especies coloreando los puntos y agrandarlos usando el argumento s
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100);
```

```python
# Debido a que seaborn usa internamente MatPlotLib, podemos usar las funciones figure y title para mejorar la gráfica
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
```

```python
# Seaborn tiene soporte integrado para dataframes de Pandas. Así que en lugar de pasar cada columna como una serie
#   se puede proporcionar el nombre de columna y usar el argumento data para especificar un dataframe
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=100,
                data=flowers_df);
```

```python
# Histograma.- Representa la distribución de una variable creando contenedores ( intervalos ) a los largo del rango de valores
#              y mostrando barras verticales para indicar el número de observaciones en casa contenedor.
```

```python
# Por ejemplo, visualicemos la distribución de valores de ancho de sépalo en el conjunto de datos Irir.
#  se puede usar la función hist para crear un histograma
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")

flowers_df.sepal_width
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

```python
# Podemos notar que los anchos de séṕalos se encuentran en el rango de 2.0 y 4.5, y que alrededor de 25 valores están 
# dentro del rango de 2.9 a 3.1, que parece ser el contenedor mas poblado.
```

```python
# Podemos controlar el tamaño y cantidad de contenedores, así como el tamaño de cada uno usando el argumento bins
# Specifying the number of bins
plt.hist(flowers_df.sepal_width, bins=5);
```

```python
import numpy as np

# Specifying the boundaries of each bin
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
# Bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

```python
# También es posible dibujar varios histogramas en un solo gráfico y reducir la opacidad para que no oculten una a la otra.
# Por ejemplo dibujar histogramas separados para cada especie 
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']

plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
```

```python
# También es posible apilar los histogramas uno encima del otro
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

```python
# Otro Dataframe incluido en seaborn es el llamado "tips" ( propinas )
tips_df = sns.load_dataset("tips");

tips_df
```

```python
# Podemos graficar como varía el monto promedio de la factura en cada día de la semana
# debido a que los promedios es un caso muy común, seaborn proporciona "barplot" que lo calcula automáticamente
sns.barplot(x='day', y='total_bill', data=tips_df);
```

```python
# Las líneas negras sobre las barras representan la cantidad de variación en los valores
# La variación es alta los viernes y baja los sábados
```

```python
# Es posible especificar el argumento "hue" para realizar comparaciones con una tercera categoría como el género
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df);
```

```python
# Se puede cambiar a barras horizontales cambiando los ejes.
sns.barplot(x='total_bill', y='day', hue='sex', data=tips_df);
```

```python
# Heatmaps ( mapas de calor ). Se usa para visualizar datos bidimensionales como una matriz o una tabla usando colores
# Usarémos el dataframe de SeaBorn llamado "flights" que muestra el paso mensual de pasajeros en un aeropuerto durante 12 años
flights_df = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

flights_df
```

```python
# Es una matríz con una fila para cada mes y una columna para cada año. Los valores muestran el número de pasajeros ( en miles )
#   que visitaron el aeropuerto en un mes específico del año. Podemos usar la función "heatmap" para visializar esto
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df);
```

```python
# los colores mas brillantes muestran donde hay mayor número de pasajeros.
# En Julio y agosto duele haber mas afluencia
# El numero de pasajeros en casa mes suele crecer año con año.
```

```python
# También podemos mostrar los valores especificando annot=True y usando el argumento "cmap" para cambiar el color
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True, cmap='Blues');
```

```python
# Podemos desactivar los ejes y las líneas de cuadrícula y mostrar el título usando las funciones corectas
plt.grid(False)
plt.title('A data science meme')
plt.axis('off')
plt.imshow(img);
```

```python
# Renderizar múltiples gráficos en una cuadrícula con MatPlotLib y Seaborn
# Esto se logra usando "subplots" que regresa un conjunto de ejes para trazar
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Use the axes for plotting
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')


# Pass the axes into seaborn
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0,1]);

# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

# Plot an image using the axes
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
```
